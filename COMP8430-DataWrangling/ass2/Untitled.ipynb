```python
import pandas as pd
emp_edu_data = pd.read_csv('emp-edu.csv')
medical_data = pd.read_csv('medical.csv')
```

```python
new_emp_edu_data = emp_edu_data[['ssn', 'gender', 'birth_date', 'education', 'occupation', 'salary', 'years_of_experience', 'employment_timestamp']]
new_medical_data = medical_data[['ssn', 'height', 'weight', 'blood_pressure', 'cholesterol_level', 'consultation_timestamp']]
```

```python
new_emp_edu_data = new_emp_edu_data.drop_duplicates()
new_emp_edu_data = new_emp_edu_data[new_emp_edu_data.birth_date.str.split('/', expand=True)[1].astype(int) < 13]
new_emp_edu_data = new_emp_edu_data.dropna()
# ssn = new_emp_edu_data['ssn']
# new_emp_edu_data[ssn.isin(ssn[ssn.duplicated()])].sort_values('ssn')
# new_emp_edu_data['ssn'].value_counts()
len(new_emp_edu_data)
```

```python
new_medical_data = new_medical_data.drop_duplicates()
new_medical_data = new_medical_data.dropna()
new_medical_data = new_medical_data[new_medical_data.height > 0]
new_medical_data = new_medical_data[new_medical_data.weight > 0]
ssn = new_medical_data['ssn']
new_medical_data[ssn.isin(ssn[ssn.duplicated()])].sort_values('ssn')
# print(len(new_medical_data))
# len(set(medical_data.ssn))
```

```python
ssn_id = set(new_emp_edu_data['ssn']) & set(new_medical_data['ssn'])
new_emp_edu_data = new_emp_edu_data[new_emp_edu_data.ssn.isin(ssn_id)]
res = pd.merge(new_emp_edu_data, new_medical_data, on='ssn')
# res.isnull().values.any()
res = res[abs(res.employment_timestamp.str.split('-', expand=True)[0].astype(int) - res.consultation_timestamp.str.split('-', expand=True)[0].astype(int)) < 5]
```

```python
res.to_csv('result.csv')
```
